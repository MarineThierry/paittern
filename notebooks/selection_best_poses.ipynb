```python
keypoints_test = np.random.random((47,17,3))
keypoints_test
```

```python
keypoints_test[8]
```

```python
distance_shoulder = []
horizontal_shoulder = []
horizontal_elbow = []
horizontal_wrist = []
horizontal_st_deviation = []

for image in keypoints_test:                                       #iterate on each image
    distance_shoulder.append(abs(image[5][0]-image[6][0]))         #calculate abs distance (on x) between 'left_shoulder', 'right_shoulder'
    horizontal_shoulder.append(abs(image[5][1]-image[6][1]))       #calculate abs distance (on y) between 'left_shoulder', 'right_shoulder'
    horizontal_elbow.append(abs(image[7][1]-image[8][1]))          #calculate abs distance (on y) between  elbows  
    horizontal_wrist.append(abs(image[9][1]-image[10][1]))         #calculate abs distance (on y) between  wrists
    

    
max_distance_x_index = distance_shoulder.index(max(distance_shoulder))         #index of image for max distance  on x shoulder
min_distance_y_index = horizontal_shoulder.index(min(horizontal_shoulder))     #index of image for min distance  on y shoulder
```

```python
list_shoulderx_sort = sorted(distance_shoulder, reverse=True)
list_shoulderx_sort
```

```python
#image index considering ranking on shoulders max(x)
image_index_ranking_x = []
for i in list_shoulderx_sort:
    image_x_index = distance_shoulder.index(i)
    image_index_ranking_x.append(image_x_index)
image_index_ranking_x 
```

```python
image_index_ranking_x_top10 = image_index_ranking_x[0 :9 ]
image_index_ranking_x_top10
```

```python
keypoints_test_top10 = []
for i in image_index_ranking_x_top10:
    keypoints_test_top10.append(keypoints_test[i])
keypoints_test_top10
    
```

```python
horizontal_shoulder_top10 = []
for image in keypoints_test_top10:                                 #iterate on each image
    horizontal_shoulder_top10.append(abs(image[5][1]-image[6][1]))       #calculate abs distance (on y) between 'left_shoulder', 'right_shoulder'
horizontal_shoulder_top10 = sorted(horizontal_shoulder_top10, reverse=False)
horizontal_shoulder_top10
```

```python
#image index considering ranking on shoulders min(y)
image_index_rank_max_x_min_y = []
for i in horizontal_shoulder_top10:
    image_index_rank_max_x_min_y.append(horizontal_shoulder.index(i))
image_index_rank_max_x_min_y
```
